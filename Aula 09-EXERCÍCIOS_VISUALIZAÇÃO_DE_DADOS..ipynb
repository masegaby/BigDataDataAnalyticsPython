#### Exercício 01 - Leia o lucro total de todos os meses e mostre-o usando um gráfico de linhas.
#### Arquivo: company_sales_data.csv
#### Dados de lucro total fornecidos para cada mês. O gráfico de linha gerado deve incluir as seguintes propriedades:
#### • Nome do rótulo X = Número do mês
#### • Nome do rótulo Y = lucro total


```python
import pandas as pd
import matplotlib.pyplot as plt

# Mapeando o caminho do arquivo:
caminho='./Arquivos/company_sales_data.csv'

#Alocar os dados em uma dataframe
df= pd.read_csv(caminho)

#Visualizar cabeçalho do dataframe
df.head()
```

```python
#Selecionar os dados que serão apresentados nos gráfico 

listaLucro= df['total_profit'].to_list()
listaMeses= df['month_number'].to_list()

print(listaLucro)
print(listaMeses)

#Criando o gráfio utilizando os dados da lista anterior

plt.plot(listaMeses,listaLucro, label= 'Lucro Mensal do Ano Anterior',c='hotpink')

plt.xlabel('Mês') #Titulo do eixo X
plt.ylabel('Lucro') #Titulo do eixo Y
plt.title('Lucro Mensal') #Titulo do gráfico
plt.legend() #Legenda do gráfico
```

#### Exercício 02 - Obtenha o lucro total de todos os meses e mostre o gráfico de linhas com as seguintes propriedades de estilo.
#### O gráfico de linha gerado deve incluir as seguintes propriedades de estilo:
#### • O estilo da linha pontilhada e a cor da linha devem ser vermelhas
#### • Mostrar legenda no local inferior direito.
#### • Nome do rótulo X = Número do mês
#### • Nome do rótulo Y = número de unidades vendidas
#### • Adicione um marcador de círculo.
#### • Cor do marcador de linha conforme lida
#### • A largura da linha deve ser 3


```python

plt.plot(listaMeses,
         listaLucro, 
         label= 'Lucro Mensal do Ano Anterior', #Legenda
         c='red',
         ls= '--', #Tipo de linha
         marker= 'o',
         mec='black',        
         markerfacecolor='k', #Remove a borda do marcador
         linewidth=3 #Espessura da linha, 
         )

plt.xlabel('Meses') #Titulo do eixo X
plt.ylabel('Vendas em dólares') #Titulo do eixo Y
plt.title('Vendas por mês no ano anterior') #Titulo do gráfico
plt.legend() #Legenda do gráfico
plt.yticks([100000,200000,300000,400000,500000])  #Ajustando os intervalos de Y
plt.xticks(listaMeses) #Ajustando os r[otulos do eico X]
plt.legend(loc='lower right') #Posicionamento da legenda)
```

#### Além disso, adicione uma grade no gráfico. o estilo da linha de grade deve “–“. O gráfico de dispersão deve ficar assim.



```python
#Importação das bibliotecas:
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#Mapear caminho do arquivo 
caminho='./Arquivos/company_sales_data.csv'

#Alocar os dados em data frame
df= pd.read_csv(caminho)
df.head()
#Gerar lista
listaMeses= df['month_number'].to_list()
vendasCremeDental= df['toothpaste'].to_list()

#Criar o gráfico
sns.scatterplot(x=listaMeses, y=vendasCremeDental,c='#7B31F5')

#ADicionar grade ao gráfico 
plt.grid(True,linewidth= 0.5,linestyle='--',c='#E2DDF4')

#ADicionar título ao grádico 
plt.title('Vendas de creme dental',c='#F531AE')
plt.xlabel('Meses')
plt.ylabel('Unidades vendidas')
```

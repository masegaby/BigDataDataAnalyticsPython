### m) Gere um arquivo em Excel para representar as vendas efetuadas pela filial Sudeste em 2010. Sendo que no relatório deve conter apenas vendas de casas sem garagem.

```python
df_final=df[(df['Filial']== 'Sudeste')&
            (df['Data'] >='2010-01-01')&
            (df['Data']<= '2010-12-31')&
            (df['Tipo']== 'Casa')&
            (df['Garag.']== 0)]

caminho= 'C:\\Users\\FIC\\Documents\\gabrielly_holanda_big_data\\Relatórios Filiais\\'
arquivo= 'Sudeste_2010_Casas_semgaragem.xlsx'

#Salvando sem ID
df_final.to_excel(caminho+arquivo,index=False)
```

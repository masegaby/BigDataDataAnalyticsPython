```python
#Renomeando colunas
df= df.rename(columns={
    'Coluna 1':'Data',
    'Coluna 2':'Vendedor',
    'Coluna 3':'Cargo',
    'Coluna 4':'Cliente',
    'Coluna 5':'UF',
    'Coluna 6':'Vendas'
})

df.head()
```

```python
#Crie um programa para criar um dataframe para mostrar as vendas por cargo, e salvar um arquivo em csv para com o nome de cada cargo:

#Filtrando dados de uma forma mais fácil 
# 1) Selecionar coluna de interesse
df_cargos= df['Cargo'].drop_duplicates()

# 2) Converter o df_cargos em lista
cargos= df_cargos.to_list()
#cargos

# 3) Filtrar o dataframe utilizando a lista convertida
for cargo in cargos:
   
    df_final= df[df['Cargo']==cargo]

    caminho = './Relatórios Exercício 3/'
    arquivo= cargo + '.xlsx'

    df_final.to_excel(caminho+arquivo)
```

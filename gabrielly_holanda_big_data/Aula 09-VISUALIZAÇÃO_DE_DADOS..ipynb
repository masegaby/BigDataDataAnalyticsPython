#### Primeiros passos com Matplotlib

```python
#Importando bibliotecas
import matplotlib.pyplot as plt
import numpy as np

#Criando as arrays (matrizes) para utilizar nos gráficos
x= np.array([1,2,3,4,5])
y= np.array([2,4,6,8,10])

#Criando um gráfico com os valores anteriores
plt.plot(x,y)
```

#### Modificando o tipo de marcador

```python
plt.plot(x,y,marker='*', c='pink')
```

#### Modificando a forma como os pontos são ligados e o tipo de linha

```python
plt.plot(x,
         y,
         marker='*',
         c='red',
         ds='steps', #Forma de ligação
         ls= '--' #Estilo da linha
         )
```

### Alterando as cores dos marcadores

```python
plt.plot(x, 
         y,
         marker='*', #Marcador
         c='#31EBFF',#Cor da linha
         ds='steps', #Forma de ligação
         ls= '--', #Estilo da linha
         mec= '#FF1DE5' #Cor dos marcadores
         )
```

#### Adicionando outros elementos ao gráfico

```python
plt.plot(x,y) #Criando gráfico

#Adicionando  o título do eixo x
plt.xlabel('Valores de X', fontsize=12)

#Adicionand o título do eixo y
plt.ylabel('Valores de Y', fontsize=12)

#Adicionando o título do g´rafico 
plt.title('Título do Gráfico', fontsize=20)

#Alterando as cores do rótulos de X e Y
plt.xticks(fontsize=8,c='blue')
plt.yticks(fontsize=8,c='purple')
```

# Multiplas curvas no mesmo mecanismo gráfico

```python
import numpy as np    

x= np.linspace(0,2*np.pi,100)
y= np. sin(x)
z= np.cos(x)

plt.plot(x,y,label ='FUNÇÃO SENO',c= '#FF19B0')
plt.plot(x,z, label= 'FUNÇÃO COSSENO', color='#637BFF')

#ADICIONAR LEGENDA AO GRÁFICO 
plt.legend()

#ADICIONAR TITULO AO GRÁFICO 
plt.title('Funções Trigonométricas')

#ADICIONAR TITULO AOS EIXOS
plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')

#SALVAR O GRÁFICO COMO IMAGEM 
plt.savefig('./Gráficos/')
```

# EXEMPLO: GRÁFICO DE PIZZA

```python
# I) IMPORTAR A BIBLIOTECA
import matplotlib.pyplot as plt

# II) DADOS PARA CRIAR GRÁFICO
explode= [0.1,0,0,0,0,0]

#Defiindo cores:
cores=['#04D939','#03A63C','#027333','#025959','#012340','#10403B']
languages= ['Java','Python', 'PHP','JavaScript', "C#","C++"]
popularidade= [22.2,17.6,8.8,8,7.7,6.7]

# III) GERAR GRÁFICO 

plt.pie(popularidade,labels= languages,autopct='%1.2f%%',shadow=True,explode= explode,colors=cores)


#Explosão das fatias 

#labels= rótulo de categorias 
#autopct= rótulos numérico %1.2= quantidade de casas decimais
#shadow= sombra
#explode= afastar umas das fatias da pizza (depende da ordem que escolheu dentro da lista)



```

### Gráfico de dispersão 

```python
#Importando as bibliotecas

import seaborn as sns
import matplotlib.pyplot as plt
from pylab import randn #Busca uma informação em epecifico dentro da biblioteca

#Geraçãio de dados aleatórios
x= randn(200)
y= randn(200)

#Criando o gráfico de dispersão
sns.scatterplot(x=x,y=y)
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

caminhoArquivo= ('./Arquivos/sea-level.csv')
df = pd.read_csv(caminhoArquivo)

#Fatiar o dataframe (selecionar a partir da linha 5 (5:) e todas as colunas(:))

df= df.loc[6:,:]

df.columns=['Year',
            'CSIRO - Adjusted sea level (inches)',
            'CSIRO - Lower error bound (inches)',
            'CSIRO - Upper error bound (inches)',
            'NOAA - Adjusted sea level (inches)'
            ]



df['Year']= df['Year'].astype(int)
df['CSIRO - Adjusted sea level (inches)']= df['CSIRO - Adjusted sea level (inches)'].astype(float)
df['CSIRO - Lower error bound (inches)']= df['CSIRO - Lower error bound (inches)'].astype(float)
df['NOAA - Adjusted sea level (inches)']= df['NOAA - Adjusted sea level (inches)'].astype(float)

anos= df['Year'].to_list()
nivel= df['CSIRO - Adjusted sea level (inches)'].to_list()


sns.scatterplot(x= anos, y = nivel)
plt.xlabel('Year')
plt.ylabel('CSIRO - Adjusted sea level (inches)')
```
